Code used to download daily stock price using python package yfinance. 
A file was created as a csv to store downloaded price on the initial run.
After initial run, only new data are downloaded, then append to the existing data. 

```python
!pip install yfinance
import yfinance as yf
import numpy as np
import datetime as dt
import bs4 as bs
from datetime import datetime

import requests
import pandas as pd
from pandas import DataFrame
import pandas_datareader.data as web
import pandas_datareader as web
from concurrent import futures
```

```python
# Get Company Information (i.e. Sector, Industry)
def save_sp500_tickers():
    resp = requests.get('http://en.wikipedia.org/wiki/List_of_S%26P_500_companies')
    soup = bs.BeautifulSoup(resp.text, 'lxml')
    table = soup.find('table', {'class': 'wikitable sortable'})

    tickers = []
    securities = []
    industries = []
    subindustries = []
    headquarters = []
    added = []
    founded = []

    for row in table.findAll('tr')[1:]:
        ticker = row.findAll('td')[0].text
        security = row.findAll('td')[1].text
        industry = row.findAll('td')[3].text
        subindustry = row.findAll('td')[4].text
        headquarter = row.findAll('td')[5].text
        add = row.findAll('td')[6].text
        found = row.findAll('td')[8].text
  
        tickers.append(ticker)
        securities.append(security)
        industries.append(industry)
        subindustries.append(subindustry)
        headquarters.append(headquarter)
        added.append(add)
        founded.append(found)

        tickers = list(map(lambda s: s.strip(), tickers))
        securities = list(map(lambda s: s.strip(), securities))
        industries = list(map(lambda s: s.strip(), industries))
        subindustries = list(map(lambda s: s.strip(), subindustries))
        headquarters = list(map(lambda s: s.strip(), headquarters))
        added = list(map(lambda s: s.strip(), added))
        founded = list(map(lambda s: s.strip(), founded))

        tickerdf = pd.DataFrame(tickers, columns = ['ticker'])
        securitydf = pd.DataFrame(securities,columns = ['security'])
        sectordf = pd.DataFrame(industries, columns = ['industry'])
        subsectordf = pd.DataFrame(subindustries, columns =['subindustry'])
        headquarterdf = pd.DataFrame(headquarters, columns = ['headquarters'])
        adddf = pd.DataFrame(added, columns = ['added'])
        founddf = pd.DataFrame(founded, columns = ['founded'])

        tickerandsecurityandsectorandsubsectorandheadquarterandaddandfound = pd.concat([tickerdf, securitydf, sectordf, subsectordf, headquarterdf, adddf, founddf], axis = 1).reindex(tickerdf.index)
        data = tickerandsecurityandsectorandsubsectorandheadquarterandaddandfound
    
    print(data)
    df = data
    df.to_csv('c://Users/JCW/Desktop/covidStocks/SP500CompanyInfo.csv')
    data.to_pickle(r'sp500_data.pickle')
save_sp500_tickers()
```

Clean up downloaded data. 
Today's data isn't always correct, so code below will remove today's data.
Sometimes, duplicate data are downloaded, code below will remove duplicates as well.

```python
## Clean up downloaded data. 
if len(df_stocks)>0: 
    df_stocks=df_stocks.reset_index()
    df_stocks=df_stocks[df_stocks['Date']<pd.to_datetime('today').floor('d')]
if len(df_stocks)>0:
    df_stocks['Total']=df_stocks['High']+df_stocks['Volume']
    df_stocks=df_stocks[df_stocks['Total'] == df_stocks.groupby(['name', 'Date'])['Total'].transform('max')]
    df_stocks=df_stocks.drop('Total', axis=1)
    df_stocks=df_stocks.drop_duplicates()
    df_stocks=df_stocks.set_index('Date')
    stocks=stocks.set_index('Date')
    stocks=stocks.append(df_stocks, sort=False, ignore_index=False)      
    stocks.to_excel('c:/Users/JCW/Desktop/covidStocks/covidStocks2.xlsx', sheet_name='covidStocks')
```
